# Phase 1 - Data Cleaning

## Step 1.1 Importing the Libraries

```python
import pandas as pd
import numpy as np
```

## Step 1.2 Loading the data

```python
df = pd.read_csv("data/raw_amazon_saas_sales.csv")
```

```python
df.head()
```

## Step 1.3 Converting the Date columns
### Here the dataset has:
- Order Date
- Date Key (YYYYMMDD)

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

## Step 1.4 Creating order_month
### We need monthly revenue, not daily.

```python
df['order_month'] = df['Order Date'].dt.to_period('M').dt.to_timestamp()
```

## Step 1.5 Selecting only the useful columns

```python
df = df[
    [
        'Customer ID',
        'Customer',
        'order_month',
        'Region',
        'Segment',
        'Industry',
        'Product',
        'Sales'
    ]
]
```

```python
df.info()
df.head()
```

## Step 1.6 Exporting the cleaned data
### We export the cleaned data into cleaned_sales.csv file

```python
df.to_csv("data/cleaned_sales.csv", index=False)
```
